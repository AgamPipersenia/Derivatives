```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_payoff(asset_prices, profit_loss, strategy_name):
    plt.plot(asset_prices,profit_loss)
    plt.grid(True)
    plt.title(strategy_name)
    plt.show()
```

```python
asset_price = np.linspace(50,150,100)
```

```python
def long_call(asset_price, strike_price, premium):
    total_pl = -premium + np.maximum(asset_price - strike_price,0)
    return total_pl

long_call_pl = long_call(asset_price, 100, 5)
plot_payoff(asset_price,long_call_pl,"long call")
```

```python
def short_call(asset_price, strike_price, premium):
    total_pl = premium - np.maximum(asset_price-strike_price,0)
    return total_pl

short_call_pl = short_call(asset_price, 100, 5)
plot_payoff(asset_price,short_call_pl,"short call")
```

```python
def long_put(asset_price, strike_price, premium):
    total_pl = -premium + np.maximum(strike_price-asset_price,0)
    return total_pl

long_put_pl = long_put(asset_price,100,5)
plot_payoff(asset_price,long_put_pl,"long put")
```

```python
def short_put(asset_price, strike_price, premium):
    total_pl = premium - np.maximum(strike_price-asset_price,0)
    return total_pl

short_put_pl = short_put(asset_price, 100, 5)
plot_payoff(asset_price, short_put_pl, "short put")
```
